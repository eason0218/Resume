# 特定學校及學制101-105年學生人數變化

```python
data2 = data
data2
```

```python
all_college = pd.DataFrame(data2.groupby(['學年度','學校名稱','學制'])['總計'].sum()).reset_index()
all_college
```

```python
nutc = all_college[all_college['學校名稱'] == '國立臺中科技大學']
nutc.head(5)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar


d = (
    Bar()
    .add_xaxis(['二專(日間)','二專(進修)','五專','學士班(日間)','學士班(進修)','碩士在職專班','碩士班(日間)'])
    .add_yaxis("101", list(nutc[nutc['學年度']==101]['總計']))
    .add_yaxis("102", list(nutc[nutc['學年度']==102]['總計']))
    .add_yaxis("103", list(nutc[nutc['學年度']==103]['總計']))
    .add_yaxis("104", list(nutc[nutc['學年度']==104]['總計']))
    .add_yaxis("105", list(nutc[nutc['學年度']==105]['總計']))
    .set_global_opts(title_opts=opts.TitleOpts(title="臺中科技大學101-105學生人數", subtitle="各學制"))
    .render_notebook()
)
d
```
