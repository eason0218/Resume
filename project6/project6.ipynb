```python
df1 = df1.drop(0,axis=0).reset_index(drop=True)
df2 = df2.drop(0,axis=0).reset_index(drop=True)
df3 = df3.drop(0,axis=0).reset_index(drop=True)
df4 = df4.drop(0,axis=0).reset_index(drop=True)
```

# 目標1刪除住宿用途外其他資料

```python
df_concat = df_concat[['鄉鎮市區', 'city', '主要用途', '都市土地使用分區', '總價元', '建物移轉總面積平方公尺', '主建物面積', '交易年月日',
                      '單價元平方公尺','建物現況格局-房', '建物現況格局-廳', '建物現況格局-衛']]
df_concat.head(5)
```

```python
columns_mapping = {'鄉鎮市區':'towns', 'city':'city', '主要用途':'main_use', '都市土地使用分區':'use_zoning'
                  ,'總價元':'total_price', '建物移轉總面積平方公尺':'building_area_square_meter', '主建物面積': 'main_building_area'
                  ,'交易年月日':'tx_dt', '單價元平方公尺':'unit_price', '建物現況格局-房':'room_number', 
                    '建物現況格局-廳':'hall_number', '建物現況格局-衛':'health_number'}
columns_type = {'total_price': 'int','unit_price':'float','building_area_square_meter':'float',
                                      'main_building_area': 'float',
                                      'room_number': 'int','hall_number': 'int','health_number': 'int'}
```

```python
df_concat = df_concat.rename(columns=columns_mapping)
df_concat.head(5)
```

```python
df_concat = df_concat.loc[(df_concat['main_use'] == '住家用')& (df_concat['use_zoning']=='住')].dropna()
```

```python
df_concat.info()
df_concat = df_concat.astype(columns_type)
df_concat.info()
```

# 目標2 新增年份並限制109年以及滿足房、廳條件

```python
df_concat['tx_dt_year'] = df_concat['tx_dt'].apply(lambda x: int(x[:-4]))
df_concat = df_concat.loc[(df_concat.tx_dt_year>=109)&
                                  (df_concat.room_number>=1)&
                                  (df_concat.room_number<=5)&
                                  (df_concat.hall_number>=1)&
                                  (df_concat.hall_number<=2)]
df_concat = df_concat.reset_index(drop=True)
df_concat
```

# 目標3透過平方公尺計算出坪

```python
# 1. 建物移轉總面積坪(building_area_square_feet) : 建物移轉總面積平方公尺*0.3025
# 2. 主建物面積坪(main_building_area_square_feet) : 主建物面積*0.3025
# 3. 單價元坪(unit_price_square_feet) : 單價元平方公尺/0.3025
df_concat['building_area_square_feet'] = df_concat.building_area_square_meter*0.3025
df_concat['main_building_area_square_feet'] = df_concat.main_building_area*0.3025
df_concat['unit_price_square_feet'] = df_concat.unit_price/0.3025
```
